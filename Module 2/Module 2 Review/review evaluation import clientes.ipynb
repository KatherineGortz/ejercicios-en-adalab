```python
df_clientes = pd.read_csv('clientes.csv', header=None)
df_clientes.columns = ['nombre', 'apellido', 'direccion', 'telefono', 'email', 'fecha_registro']


```

```python
query = """
    INSERT INTO clientes (nombre, apellido, direccion, telefono, email, fecha_registro)
    VALUES (%s, %s, %s, %s, %s, %s)   
"""

lista_clientes = []

for client, row in df_clientes.iterrows():
    nombre = row["nombre"]
    apellido = row["apellido"]
    direccion = row["direccion"]
    telefono = row["telefono"]
    email = row["email"]
    fecha_registro = row["fecha_registro"]

    lista_clientes.append((nombre, apellido, direccion, telefono, email, fecha_registro))

mycursor.executemany(query, lista_clientes)

cnx.commit()
```
