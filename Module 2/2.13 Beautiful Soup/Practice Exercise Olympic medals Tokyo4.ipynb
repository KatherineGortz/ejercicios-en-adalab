```python
from bs4 import BeautifulSoup
import requests

# Importar librerías para manipulación y análisis de datos
# -----------------------------------------------------------------------
import pandas as pd

# Importar librerías para procesamiento de texto
# -----------------------------------------------------------------------
import re
```

```python
# definimos la url de la página de la vamos a sacar datos
url_medallas = "https://es.wikipedia.org/wiki/Anexo:Medallero_de_los_Juegos_Ol%C3%ADmpicos_de_Tokio_2020"

# hacemos la request a la página de la que queremos sacar la info
res_medallas = requests.get(url_medallas)

# vemos si todo ha ido bien
print("La respuesta de la petición es:", res_medallas.status_code)
```

```python
# creamos el objeto BeautifulSoup para poder acceder al contenido solicitado
sopa_medallas = BeautifulSoup(res_medallas.content, 'html.parser')

# mostramos por pantalla los resultados del objeto de Beautiful Soup. El método ".prettify()" nos ayuda a visualizar los resultados de una forma más amigable
print(sopa_medallas.prettify())
```

```python
# sacamos los nombres de los moviles
lista_medallas = sopa_medallas.find_all("table")
print("El resultado del método '.find_all()' es: \n", lista_medallas)

# "table wikitable striped sortable col1cen col2izq col3der col4der col5der col6der jquery-tablesorter"
```

```python
print("El número de tablas que tenemos en la página web es:", len(lista_medallas))
```

```python
print("En la tabla 1 tenemos: \n ", lista_medallas[0])
```

```python
tabla_html_medallero = lista_medallas[0]
```

```python
lista_encabezados = tabla_html_medallero.find_all("th")

print("La lista que nos devuelve el metodo '.find_all()' es:\n", lista_encabezados)

# como hemos estado haciendo hasta ahora, tendremos que iterar por la lista obtenida en el paso anterior y extraer el texto de cada elemento
encabezados_medallas = [
    th.find("a")["title"] if th.find("a") and th.find("a").has_attr("title") 
    else th.get_text(strip=True)
    for th in lista_encabezados
]

print("\n-------------------------\n")
print("Los encabezados de la tabla son:", encabezados_medallas)
```

```python
filas_medallas = tabla_html_medallero.find_all("tr")
print("La cantidad de filas que hemos extraido de la tabla es:", len(filas_medallas))
```

```python
print("\n----------------------\n")
print("El contenido que tenemos en la primera fila es:\n", filas_medallas[0])

print("\n----------------------\n")
print("El contenido que tenemos en la segunda fila es:\n", filas_medallas[1])
```

```python
medallas_por_pais = []

for fila in filas_medallas[1:]:
    # Para cada 'fila', extraemos el texto, lo dividimos en una lista usando '\n' como separador y eliminamos el primer y último elemento de la lista.
    fila_texto = fila.text
    elementos_fila = fila_texto.split("\n")[1:-1]
    medallas_por_pais.append(elementos_fila)

print (medallas_por_pais)
```

```python
# Iteramos a través de cada lista en 'resultados_ibex' y sus elementos usando 'enumerate'.
for indice_lista, lista in enumerate(medallas_por_pais):
    for indice_elemento, elemento in enumerate(lista):
        try:
            # Intentamos convertir cada elemento de la lista en un número de punto flotante.
            # Para ello, eliminamos caracteres especiales como '%' y reformateamos la puntuación.
            medallas_por_pais[indice_lista][indice_elemento] = elemento.replace("\xa0", " ")
        except:
            # Si se produce una excepción (por ejemplo, si el elemento no es convertible a flotante), 
            # mantenemos el elemento original en la lista.
            medallas_por_pais[indice_lista][indice_elemento] = elemento
            
# Imprimimos los resultados después de limpiar los datos.
print("Los resultados después de limpiar los datos son: \n", medallas_por_pais)
```

```python
# convertimos la lista de listas a DataFrame usando el método 'pd.DataFrame()'. 
df_medallas = pd.DataFrame(medallas_por_pais)
```
