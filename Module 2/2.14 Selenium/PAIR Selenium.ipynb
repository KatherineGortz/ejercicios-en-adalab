```python
##SOlucion Eleonora SIN BACK SIN CLICK

url = "https://www.casadellibro.com"
driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()
sleep(5)
driver.find_element("xpath", '//*[@id="onetrust-accept-btn-handler"]').click()
sleep(2)
driver.find_element('link text', "Ficción").click()
sleep(2)
driver.find_element('link text', "Contemporánea").click()

diccionario_libros_contemporáneos2= {"Título del libro": []} 

for i in range(1,21):
    sleep(5)
    try: 
        driver.find_element('xpath', f'//*[@id="App"]/div[3]/cma-component[5]/div/div[3]/div[3]/div[{i}]')
        sleep(3)
        #"Título del libro"
        try:
            titulo  = driver.find_element('xpath', f'//*[@id="App"]/div[3]/cma-component[5]/div/div[3]/div[3]/div[{i}]/div[1]/a').text
        except:
            titulo= "No info"
        diccionario_libros_contemporáneos2["Título del libro"].append(titulo)

    except Exception as e:
        print(f"El errror ha sido {e}")
        driver.back()

driver.find_element('xpath', '//*[@id="App"]/div[3]/cma-component/div/div[3]/div[4]/div/button[1]')

```
