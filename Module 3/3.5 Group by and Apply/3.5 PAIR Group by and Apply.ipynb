1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("world_data_full.csv", index_col = 0)
```

```python
def cambiar_dtype(cadena):
    try:
    # Reemplazar las comas por puntos en la cadena
        return float(cadena.replace(",", ".").replace(" ", ""))

    except:
        # Si ocurre algún error (por ejemplo, si el argumento no es una cadena),
        # devolver np.nan (valor Not a Number, vamos, un NULL de toda la vida)
        return np.nan
```

```python
lista_columnas = ['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitude', 'longitude', 'density', 'cpi']


for col in lista_columnas:
    df[col] = df[col].apply(cambiar_dtype)
```

```python
df['gdp']=df['gdp'].astype(float)
```

2. Crea una función para añadir una nueva columna llamada 'Continent' al DataFrame para indicar a qué continente pertenece cada país.

```python
pip install pycountry pycountry-convert
```

```python
import pycountry_convert as pc
import pycountry
```

```python
def get_continent(country):
    try:
        # Convert country name to ISO Alpha-2 code
        country_code = pycountry.countries.lookup(country).alpha_2
        # Convert country code to continent code
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        # Map continent codes to full continent names
        continents = {'AF': 'Africa', 'NA': 'North America', 'SA': 'South America', 
                      'AS': 'Asia', 'EU': 'Europe', 'OC': 'Oceania'}
        return continents.get(continent_code, 'Unknown')
    except Exception:
        return 'Unknown'  # If conversion fails

# Apply function to DataFrame
df['continent'] = df['country'].apply(get_continent)

# Print the updated DataFrame
df
```
