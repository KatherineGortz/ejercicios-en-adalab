```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
for column in categorical_columns:
    print("Valores únicos y frecuencias para", column)
    print(df[column].value_counts())
```

```python
#Seleccionad solo las columnas de tipo categórico del DataFrame. 
# Después, mostrad los valores únicos y sus frecuencias para cada una de las columnas seleccionadas.

df_objects = df.select_dtypes(include = 'object')

for col in df_objects.columns:
    print(df_objects[col].value_counts())  # Apply unique() to each column


```
