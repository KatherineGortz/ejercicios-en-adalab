```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# lo primero que hacemos es cargar el dataframe que vamos a usar para esta lección
df = pd.read_csv("merchant_fill_nonulls.csv", index_col=0) 
df.head(3)
```

```python
#¿Cómo se distribuyen los precios de los productos?

sns.histplot(data=df, x='price', bins = 25, kde=True, color='green')
```

```python
#¿Cuál es la relación entre el precio y la calificación promedio de los productos?

sns.scatterplot(data=df, x='price', y='rating_sold', size = 3, color = 'purple')
```

```python
#¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (uses_ad_boosts) o no?

sns.scatterplot(x='price', y='uses_ad_boosts', data=df)
```

```python
sns.boxplot(x= 'price', y = "uses_ad_boosts", 
            data = df, 
            width = 0.5, 
            color = "turquoise")
```

```python
sns.violinplot(x= 'price', y = "uses_ad_boosts", 
            data = df, 
            width = 0.5, 
            color = "turquoise")
```

```python
#¿Existe una relación entre el uso de anuncios y las unidades vendidas?

sns.violinplot(x='uses_ad_boosts', y='units_sold', data=df)
plt.show()
```

```python
#¿Cuál es la relación entre el precio de envío y las unidades vendidas?

sns.regplot(x='price', y='units_sold', data=df)
```

```python
#¿Cuál es la distribución de las calificaciones de los empleados (rating_employee) para los productos que tienen el distintivo "Buena calidad del producto" (badge_product_quality) y para los que no lo tienen? 
#Haced que las dos gráficas tengan la misma escala en los dos ejes.
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))



sns.violinplot(x='badge_product_quality', y='rating_employee', data=df.query("badge_product_quality == 'Buena'"), ax = axes[0])


sns.violinplot(x='badge_product_quality', y='rating_employee', data=df.query("badge_product_quality == 'Excelente'"), ax = axes[1])
```
