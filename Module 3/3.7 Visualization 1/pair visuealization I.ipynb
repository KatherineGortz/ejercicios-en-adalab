```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv") 
```

```python
df.head()
```

```python
sns.histplot(x="birth_rate", data = df);
plt.xlabel("tasa de natalidad ")
plt.ylabel("Frecuencia")
plt.title("Histograma de la tasa de natalidad")
```

```python
#"2:"
```

```python
sns.scatterplot(x="infant_mortality", y="life_expectancy", data=df, color="blue",size=2);
```

```python

df_europe= df[df['continent']=='Europe']
```

```python
sns.boxplot(x= "life_expectancy", data= df_europe, color="blue"); 
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5))
plt.subplots_adjust(wspace=0.4, hspace=0.6)
sns.regplot(x=df_europe['birth_rate'],y= df_europe['infant_mortality'], color = 'blue',ax=axes[0, 0],scatter_kws = {'s' : 2}) 
axes[0,0].set_title("birth_rate vs infant_mortality")

sns.regplot(x=df_europe['birth_rate'],y= df_europe['life_expectancy'], color = 'blue', ax=axes[0, 1])
axes[0,1].set_title("birth_rate vs life_expectancy")

sns.regplot(x=df_europe['birth_rate'],y= df_europe['density'], color = 'blue', ax=axes[1, 0])
axes[1,0].set_title("birth_rate vs density")

sns.regplot(x=df_europe['birth_rate'],y= df_europe['urban_population'], color = 'blue', ax=axes[1, 1])
axes[1,1].set_title("birth_rate vs urban_population")


```
