```python
import pandas as pd
```

```python
#1. Crea una serie con números correlativos entre 1 y 12 inclusive (puedes hacerlo como lista o diccionario):

lista_q1 = list(range(1,13))

print(lista_q1)

serie_q1 = pd.Series(lista_q1)

print(serie_q1)
```

```python
# Cambia el indice de la serie anteriormente creada para que comience con el numero 20.

serie_q1 = pd.Series(lista_q1, index = range(20, 20 + len(lista_q1)))

print(serie_q1)
```

```python
#Nuevamente cambia el indice de la serie, esta vez con una lista que haga referencia a los meses del año 
#(tenemos 12 elementos como la cantidad de meses del año).

meses_lista = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

serie_q1_meses = pd.Series(lista_q1, index=meses_lista)

print(serie_q1_meses)
```

```python
#Muestra los valores que están en la posición 3 y 8 de mi serie.
serie_q1_meses[[3, 8]]
```

```python
#Muestra los valores que están etiquetados en enero y agosto.

serie_q1_meses[['Jan', 'Aug']]
```

```python
#Muestra sus valores, indices, forma, tamaño y el tipo de elementos de la serie

print(serie_q1_meses.values)
print(serie_q1_meses.index)
print(serie_q1_meses.shape)
print(serie_q1_meses.size)
print(serie_q1_meses.dtype)
```

```python
#Cread una nueva columna donde tengamos la diferencia entre el precio al que vende el e-commerce menos el precio del minorista. 
#Insertad esta columna en la posición dos.

# Calculate the difference between 'retail_price' and 'price'
df_q2["price_diff"] = df_q2["retail_price"] - df_q2["price"]

# Insert the new column at position 2
df_q2.insert(2, "price_diff", df_q2.pop("price_diff"))
```

```python
df_q2.insert(0, 'diferencia_precios', (df_q2['price'] - df_q2['retail_price']))
```
